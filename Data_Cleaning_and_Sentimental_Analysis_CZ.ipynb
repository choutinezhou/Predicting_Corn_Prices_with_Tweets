```python
# rename columns
tweetsDF.columns=['date','retweets','favorites','text','location','language']
# tract test columns as pandas.Series
text=tweetsDF.loc[tweetsDF['language']=='en','text']
```

```python
tweetsDF.head()
```

```python
# tweets cleanser- remove url
text=text.str.replace(r"http\S+", "")
```

```python
# remove @username
text=text.str.replace(r"@http\S+", "")
```

```python
# remove symbols:
text=text.str.replace("[-:!\/_,$%^*.;()+\"\'+——！，。？、~@#￥%……&*（）]","")
# sym=['!','@','#','$','%','^','&','*','(',')']
```

```python
# remove numbers
text=text.str.replace('\d+', '')
```

```python
##????????? remove special characters like  �  (not sure how to do it)

text=text.str.replace("�","")
```

```python
analyzer = SentimentIntensityAnalyzer()
scores=[]
for t in textList:
    score = analyzer.polarity_scores(t)
    scores.append(score)
```
